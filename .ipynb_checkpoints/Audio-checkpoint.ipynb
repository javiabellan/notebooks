```python
import numpy as np
import librosa
import IPython.display as ipd # For playing audio

import matplotlib.pyplot as plt
import librosa.display
%matplotlib inline

librosa.__version__
```

## Reading Audio

```python
filename = librosa.util.example_audio_file()
x, sample_rate = librosa.load(filename)#, sr=None)

print('Clip Length   {:3.2f} seconds'.format(len(x)/sample_rate))
print('Sample Rate   {} Hz'.format(sample_rate))
```

## Visualizing 1D

```python
plt.figure(figsize=(16, 4))
librosa.display.waveplot(x, sr=sample_rate);
```

```python
x_short = x[1000:2000]
plt.plot(x_short)
plt.title('Wave with matplotlib');
```

```python
librosa.display.waveplot(x_short, sr=sample_rate);
```
