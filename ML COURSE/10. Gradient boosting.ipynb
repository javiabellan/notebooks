# Gradient boosting

- Good for litte & heterogeneous data (tabualar)
- Bad for big & homogeneous data (images, audo, text)

# Import libraries

```python
import sklearn
from sklearn import datasets
from sklearn import metrics
from sklearn.model_selection import train_test_split

from xgboost  import XGBClassifier
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
```

# Load data

```python
wine = sklearn.datasets.load_wine()

x = wine.data
y = wine.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
wine.data.shape
```

# XGBoost

```python
# fit a XGBoost model to the data
model = XGBClassifier()
model.fit(x_train, y_train)

# make predictions
expected_y  = y_test
predicted_y = model.predict(x_test)

# summarize the fit of the model
print('XGBoost: ')
print(metrics.classification_report(expected_y, predicted_y, 
               target_names = wine.target_names))
print(metrics.confusion_matrix(expected_y, predicted_y))
```

# LightGBM

```python
# fit a LightGBM model to the data
model = LGBMClassifier()
model.fit(x_train, y_train)

# make predictions
expected_y  = y_test
predicted_y = model.predict(x_test)

# summarize the fit of the model
print('LightGBM: ')
print(metrics.classification_report(expected_y, predicted_y, target_names=wine.target_names))
print(metrics.confusion_matrix(expected_y, predicted_y))
```

# CatBoost

```python
# fit a CatBoost model to the data
model = CatBoostClassifier()
model.fit(x_train, y_train)

# make predictions
expected_y  = y_test
predicted_y = model.predict(x_test)

# summarize the fit of the model
print('CatBoost: ')
print(metrics.classification_report(expected_y, predicted_y, target_names = wine.target_names))
print(metrics.confusion_matrix(expected_y, predicted_y))
```
